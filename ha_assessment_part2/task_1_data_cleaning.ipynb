# Transform Data 

```python
# Extract new dataset with empty value replaced with "unknown"
df_cleaned = df.fillna("unknown")

# create "listing_datetime" column from created_at
df_cleaned['listing_datetime'] = pd.to_datetime(df_cleaned['created_at'])

# create country column for row level context
df_cleaned['country'] = df_cleaned['country_code'].replace({'it':'Italy'})

# remove quotation mark from string values and replace non-conforming values as appropriate
df_cleaned['furnished'] = df_cleaned['furnished'].str.strip('"')
df_cleaned['furnished'] = df_cleaned['furnished'].replace({'':'unknown'})

df_cleaned['total_size'] = df_cleaned['total_size'].replace({'""':'0', '"100 over  + 35 down"':'"135"', '"-135"':'"135"', 'unknown':'0', '"80-100 m2"':'"100"'})
df_cleaned['total_size'] = df_cleaned['total_size'].str.strip('"')
df_cleaned['total_size'] = df_cleaned['total_size'].astype(int)
df_cleaned['plot_size_category'] = pd.cut(
                                        x=df_cleaned["total_size"],
                                        bins=[-np.inf, 1, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 2000, 3000, np.inf],
                                        labels=["0", "1 - 100", "101 - 200", "201 - 300", "301 - 400", "401 - 500", "501 - 600",
                                               "601 - 700", "701 - 800", "801 - 900", "901 - 1000", "1001 - 2000", "2001 - 3000", "Above 3000"],
)

df_cleaned['registration_possible'] = df_cleaned['registration_possible'].str.strip('"')
df_cleaned['registration_possible'] = df_cleaned['registration_possible'].replace({'':'unknown'})

df_cleaned['washing_machine'] = df_cleaned['washing_machine'].str.strip('"')
df_cleaned['washing_machine'] = df_cleaned['washing_machine'].replace({'':'unknown'})

df_cleaned['tv'] = df_cleaned['tv'].str.strip('"')
df_cleaned['tv'] = df_cleaned['tv'].replace({'':'unknown'})

df_cleaned['balcony'] = df_cleaned['balcony'].str.strip('"')
df_cleaned['balcony'] = df_cleaned['balcony'].replace({'':'unknown'})

df_cleaned['garden'] = df_cleaned['garden'].str.strip('"')
df_cleaned['garden'] = df_cleaned['garden'].replace({'':'unknown'})

df_cleaned['terrace'] = df_cleaned['terrace'].str.strip('"')
df_cleaned['terrace'] = df_cleaned['terrace'].replace({'':'unknown'})

# Enable access to default index column
df_cleaned.reset_index(inplace=True)

# Rename columns
df_cleaned = df_cleaned.rename(columns = {'price':'monthly_rent', 'index':'ID'})

# Extract required columns
df_cleaned = df_cleaned[['ID', 'country', 'city', 'category', 'plot_size_category', 'registration_possible', 'furnished',
                        'tv', 'washing_machine', 'garden', 'balcony', 'terrace', 'monthly_rent', 'listing_datetime']]
```
